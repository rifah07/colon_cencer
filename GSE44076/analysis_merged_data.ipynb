```python
import pandas as pd
import numpy as np
```

```python
# Load the filtered metadata file
metadata_path = "GSE44076_merged_data.csv"
metadata = pd.read_csv(metadata_path)

# replace '--' with NaN in the Stage column
metadata["stage"] = metadata["stage"].replace("--", np.nan)

# normalize stage values
metadata["stage"] = metadata["stage"].str.upper()
```

```python
stage_mapping = {
    '0': 'Early',
    'I': 'Early',
    'IA': 'Early',
    'IB': 'Early',
    'IC': 'Early',
    'II': 'Intermediate',
    'IIA': 'Intermediate',
    'IIB': 'Intermediate',
    'IIC': 'Intermediate',
    'II': 'Advanced',
    'IIIA': 'Advanced',
    'IIIB': 'Advanced',
    'IIIC': 'Advanced',
    'IV': 'Advanced',
    'IVA': 'Advanced',
    'IVB': 'Advanced',
    'IVC': 'Advanced'
}
# map the 'stage' column to the broader categories
metadata['stage_category'] = metadata['stage'].map(stage_mapping)

# handle any unmapped or missing values
metadata['stage_category'] = metadata['stage_category'].fillna('Unknown')

# Save the updated DataFrame to a new CSV file
#metadata.to_csv('processed_metadata_with_stage_categories.csv', index=False)

#print("Processed metadata saved as 'processed_metadata_with_stage_categories.csv'")
```
